```python
from torchvision.models import EfficientNet_V2_S_Weights
import torchvision.models as models

model = models.efficientnet_v2_s(weights=EfficientNet_V2_S_Weights.DEFAULT)
model.eval()
```

```python
import torch

# Crear un tensor de entrada de ejemplo (batch_size=1, 3 canales, 224x224)
dummy_input = torch.randn(1, 3, 224, 224)
```

```python
# Exportar el modelo a formato ONNX
onnx_model_path = "efficientnet.onnx"

torch.onnx.export(
    model,  # Modelo a exportar
    dummy_input,  # Tensor de entrada de ejemplo
    onnx_model_path,  # Ruta para guardar el modelo ONNX
    input_names=["input"],  # Nombre de la entrada
    output_names=["output"],  # Nombre de la salida
    dynamic_axes={"input": {0: "batch_size"}, "output": {0: "batch_size"}},
    opset_version=11,  # Versión del opset de ONNX
)

print(f"Modelo EfficientNet exportado a {onnx_model_path}")
```
